```python
#reading file data file
advertise = pd.read_csv('Advertising.csv')
```

```python
#plotting/visualising the data

#split data into predictors X and output Y
predictors =['TV']
x = advertise[predictors]
y = advertise['sales']


#split data into training and testing 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)


plt.scatter(x_train,y_train)
plt.title('Sales based on TV adverstisment')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

```python
# Initialise and fit model

lm = LinearRegression()
model =lm.fit(x,y)

print(f'alpha = {model.intercept_}')
print(f'betas = {model.coef_}')
```

```python
#Training the model visual
mymodel = model.predict(x)
plt.scatter(x,y, c='cyan')
plt.title('Sales based on TV adverstisment')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.plot(x, mymodel)
plt.show()
```

```python
lm = LinearRegression()
model =lm.fit(x_train,y_train)

print(f'alpha = {model.intercept_}')
print(f'betas = {model.coef_}')
```
